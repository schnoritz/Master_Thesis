```python
import os

patients = ["nt0","nt5","nt6","nt7","nt8","nt9","pt0","pt1","pt2"]
root_dir = "/Users/simongutwein/mnt/qb/baumgartner/sgutwein84/test_results"

prostate_gamma = []
mixed_gamma = []
for patient in patients:
    pat_dir_mixed = os.path.join(root_dir, patient, "mixed_trained_UNET_1183.pt")
    pat_dir_prostate = os.path.join(root_dir, patient, "prostate_trained_UNET_2234.pt")
    
    with open(os.path.join(pat_dir_mixed, "gamma.txt")) as fin:
        lines = fin.readlines()
        mixed_gamma.append(float(lines[-1]))
        
    with open(os.path.join(pat_dir_prostate, "gamma.txt")) as fin:
        lines = fin.readlines()
        prostate_gamma.append(float(lines[-1]))
    
    
prostate_gamma = np.array(prostate_gamma)
mixed_gamma = np.array(mixed_gamma)
```

```python
print(mixed_gamma)
print(prostate_gamma)
print(mixed_gamma - prostate_gamma)
```

```python
fig, ax = plt.subplots(1,1, figsize=(7,5)) 

size=15
ax.scatter(range(len(patients)), mixed_gamma)
ax.scatter(range(len(patients)), prostate_gamma)
ax.set_xticks(range(len(patients)))
ax.set_xticklabels(patients)#, rotation='vertical', fontsize=18)
ax.legend(["Mixed Trained", "Prostate Trained"])
for num, (pros, mix) in enumerate(zip(prostate_gamma, mixed_gamma)):
    print(num)
    if abs(mix-pros) < 2:
        pass
    elif mix > pros:
        ax.arrow(num, pros+0.5, 0, mix-pros-1.5, width=0.1, color="lightblue", head_width=0.2, head_length=1, alpha=0.5, head_starts_at_zero=False)
    else:
        ax.arrow(num, pros-0.5, 0, mix-pros+1.5, width=0.1, color="lightcoral", head_width=0.2, head_length=1, alpha=0.5, head_starts_at_zero=False)

ax.set_ylabel("Gamma-Index /%", fontsize=size)
ax.set_xlabel("Test-Patient", fontsize=size, labelpad=10)
ax.tick_params(axis='y', labelsize=size)
ax.tick_params(axis='x', labelsize=size)
plt.show()
```
